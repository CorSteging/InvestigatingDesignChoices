### Split by part

```python
med_path = 'results/parts/med/'
echrod_path = 'results/parts/echrod/'
bert_path = 'results/BERT/parts/parts.csv'
metric = 'MCC'

part_results = {}
for path_name, path in { 'echrod':echrod_path, 'med':med_path}.items():
    if path_name == 'med':
        metric = 'Accuracy'
    else:
        metric = 'MCC'
    part_results[path_name] = {}
    for part in ['facts', 'procedure', 'procedure+facts']:
        results = {}

        for classifier in os.listdir(path):
            if classifier[-3:] != 'csv': continue
            if part not in classifier: continue
            clf = classifier[8:-4]
            results[clf] = pd.read_csv(path+classifier)
            results[clf]['Classifier'] = clf.replace('_'+part, '')

        if not results: continue
        results = pd.concat(results.values())

        if False:
            table = pd.concat([
            results[results['Classifier']=='SVM_med'][['Article']],
            results[results['Classifier']=='SVM_med'][[metric]],
            results[results['Classifier']=='SVM'][[metric]],
            results[results['Classifier']=='NB'][[metric]],
            results[results['Classifier']=='RF'][[metric]],
            ],axis=1)
        else:
            table = pd.concat([
            results[results['Classifier']=='SVM_med'][['Article']],
            results[results['Classifier']=='SVM_med'][[metric]]*100,
            results[results['Classifier']=='SVM'][[metric]]*100,
            results[results['Classifier']=='NB'][[metric]]*100,
            results[results['Classifier']=='RF'][[metric]]*100,
            ],axis=1)

        table.columns = ['Article', 'SVM (rep)', 'SVM (SGD)', 'Naive Bayes', 'Random Forest']
        table = table.round(decimals=2).T
        part_results[path_name][part] = table
        table.to_csv('results/parts/'+path_name+'_'+part+'.csv')
```

# Temporal effects

```python
metric = 'MCC' #'accuracy'
plt.rcParams['figure.dpi'] = 150
year_results = {}
window_results = {}
random_results = {}
for clf in ['BERT', 'SVM', 'NB', 'RF']:
    for name, df in {'only': year_results, 'random': random_results, 'window': window_results}.items():
        df[clf] = pd.read_csv('results/'+clf+'/years/years_'+name+'.csv')

        # Drop all cases before 1978 because most contain too little data
        df[clf] = df[clf][df[clf]['year'] >= 1978]
        df[clf] = df[clf].sort_values(by=['year'])
        if name != 'window':
            df[clf] = df[clf].drop_duplicates(subset=['year'])
        df[clf] = df[clf].reset_index()
    
year_results['BERT'].head()
```

```python
def get_row(df, year, metric, name=None):
    grouped = df#df[df['year']==year]
    new_df = grouped[['year', metric]]
    if name:
        new_df.columns = ['year', name]
    new_df = new_df.sort_values(by=['year'])
    return new_df.drop_duplicates()

def compare_year_effect(year_results, year, metric):
    df = get_row(year_results['SVM'], year, metric, 'SVM')
    df = pd.merge(df, get_row(year_results['NB'], year, metric, 'Naive Bayes'), on='year')
    df = pd.merge(df, get_row(year_results['RF'], year, metric, 'Random Forest'), on='year')
    df = pd.merge(df, get_row(year_results['BERT'], year, metric, 'BERT'), on='year')
    df['Train size'] = year_results['SVM'].sort_values(by=['year'])['training_size'].astype(int)
    df['Test size'] = year_results['SVM'].sort_values(by=['year'])['test_size'].astype(int)
    df = df.set_index('year')
    df = round(df*100, 1)
    df['Train size'] = df['Train size'] / 100
    df['Test size'] = df['Test size'] / 100
    df = df.T
    df['Mean'] = round(df.T.mean(), 1)
    return df
```

```python
# The clfs trained on data from the past
year_df = compare_year_effect(year_results, 0, metric)
year_df.to_csv('results/by_year.csv')
year_df
```

```python
# The clfs trained on random data
random_df = compare_year_effect(random_results, 0, metric)
random_df.to_csv('results/by_year_random.csv')
random_df
```
